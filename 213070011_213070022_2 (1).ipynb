### 2(b) Training k-means, and finding the appropriate number of k.

https://www.reneshbedre.com/blog/kmeans-clustering-python.html

The code for this section is taken from above source and modified the input according to data set given.


```python
from yellowbrick.cluster import KElbowVisualizer
model = KMeans()
visualizer = KElbowVisualizer(model, k=(1,12)).fit(df) #computing no. of cluster using elbow method.
visualizer.show()
```

***Observation:*** from the above graph it an shown that based on the elbow method, the optimal number of clusters (k) is 4

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=4, init='k-means++', random_state=0).fit(df)
```

Important attributes of KMeans class ,

```python
kmeans.labels_ # Class labels
```

```python
kmeans.inertia_ #gives within-cluster sum of squares.
```

```python
kmeans.n_iter_ #number of iterations
```

```python
kmeans.cluster_centers_ #Location of the centroids on each cluster.
```

```python
from collections import Counter
Counter(kmeans.labels_) #getting the size of each cluster
```

### 2(c) Using the cluster assignment as the label, visualizing the t-sne embedding


https://www.datatechnotes.com/2020/11/tsne-visualization-example-in-python.html

The code for this part is taken from above source and modified data according to our dataset.

```python
from sklearn.manifold import TSNE

tsne = TSNE(n_components=2, verbose=1, random_state=123) #Calling tsne function 
z = tsne.fit_transform(df) #fitting the tsne function to our dataset and its result is shown below
```

```python
df = pd.DataFrame()
df["kmeans"] = kmeans # taking kmeans clustering class as label for tsne
df["comp-1"] = z[:,0]
df["comp-2"] = z[:,1]

sns.scatterplot(x="comp-1", y="comp-2", hue=df.kmeans.tolist(),
                palette=sns.color_palette("hls", 3),
                data=df).set(title=" T-SNE projection") #plotting the tsne projection
```
